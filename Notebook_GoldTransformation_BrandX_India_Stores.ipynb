### Gold Layer Transformations & Aggregations BrandX_India_Stores_Data

```python
from pyspark.sql.functions import *

brandX_df = spark.read.parquet("Files/BrandX_India_Store_Silver/BrandX_India_Stores_Data")

display(brandX_df)

brandX_df.write.mode("overwrite").parquet("Files/BrandX_India_Store_Gold/BrandX_India_Stores_Data")
```

```python
# KPI : 1 aggregates the stores data of Quanitity and revenue 
brandX_df = spark.read.parquet("Files/BrandX_India_Store_Silver/BrandX_India_Stores_Data")
brandX_df = brandX_df.groupby("Store_Name","City","Quantity_Available","Revenue").count()
display(brandX_df)
brandX_df.write.mode("overwrite").parquet("Files/BrandX_India_Store_Gold/Quantity and Revenue")
```

```python
# KPI 2: aggregates the stores data of Sales Methods a
brandX_df = spark.read.parquet("Files/BrandX_India_Store_Silver/BrandX_India_Stores_Data")
brandX_df = brandX_df.groupBy("Store_name","Sales_Channel","Payment_Method","Supplier").count()
display(brandX_df)
brandX_df.write.mode("overwrite").parquet("Files/BrandX_India_Store_Gold/Sales_Method")
```

```python
# KPI 3 : aggregates product reviews to calculate the average rating of products.abs
brandX_df = spark.read.parquet("Files/BrandX_India_Store_Silver/BrandX_India_Stores_Data")
brandX_df = brandX_df.groupBy("Product_Name").agg(avg("Online_Rating").alias("AvgRating"))
display(brandX_df)
brandX_df.write.mode("overwrite").parquet("Files/BrandX_India_Store_Gold/Product_Rating")
```
